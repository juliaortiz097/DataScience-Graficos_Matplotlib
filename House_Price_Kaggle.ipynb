```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando os dados
df = pd.read_csv('Housing.csv')
```

```python
# 1. Relação entre a área da casa e o preço
plt.figure(figsize=(8, 5))
plt.scatter(df['area'], df['price'], alpha=0.5)  # gráfico de dispersão
plt.title('Relação entre Área e Preço das Casas')
plt.xlabel('Área (m²)')
plt.ylabel('Preço')
plt.show()
```

```python
# 2. Influência do número de quartos no preço
plt.figure(figsize=(8, 5))
df.boxplot(column='price', by='bedrooms')  # gráfico de caixa por número de quartos
plt.title('Preço das Casas por Número de Quartos')
plt.suptitle('')  # remove o título superior padrão
plt.xlabel('Número de Quartos')
plt.ylabel('Preço')
plt.show()
```

```python
# 3. Variação de preço com o número de banheiros
plt.figure(figsize=(8, 5))
df.boxplot(column='price', by='bathrooms')  # gráfico de caixa por número de banheiros
plt.title('Preço das Casas por Número de Banheiros')
plt.suptitle('')
plt.xlabel('Número de Banheiros')
plt.ylabel('Preço')
plt.show()
```

```python
# 4. Preço por características especiais (exemplo com ar-condicionado)
plt.figure(figsize=(8, 5))
df.groupby('airconditioning')['price'].mean().plot(kind='bar')  # gráfico de barras
plt.title('Preço Médio por Ar-condicionado')
plt.xlabel('Ar-condicionado')
plt.ylabel('Preço Médio')
plt.show()
```

```python
# 5. Impacto de estar em uma área preferencial no preço
plt.figure(figsize=(8, 5))
df.groupby('prefarea')['price'].mean().plot(kind='bar')
plt.title('Comparação de Preço em Áreas Preferenciais')
plt.xlabel('Área Preferencial')
plt.ylabel('Preço Médio')
plt.show()
```

```python
# 6. Influência do número de andares no preço
plt.figure(figsize=(8, 5))
df.boxplot(column='price', by='stories')
plt.title('Preço das Casas por Número de Andares')
plt.suptitle('')
plt.xlabel('Número de Andares')
plt.ylabel('Preço')
plt.show()
```

```python
# 7. Correlação entre tipo de mobília e preço
plt.figure(figsize=(8, 5))
df.groupby('furnishingstatus')['price'].mean().plot(kind='bar')
plt.title('Preço das Casas por Tipo de Mobília')
plt.xlabel('Tipo de Mobília')
plt.ylabel('Preço Médio')
plt.show()
```

####A função groupby é usada para calcular médias em categorias específicas, o que é visualizado com gráficos de barras, enquanto boxplot é utilizado para mostrar a distribuição de preços por categorias como número de quartos, banheiros e andares.
